1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
# Solution
class Counter:
    def __init__(self, max_value):
        self.max_value = max_value
        self.value = 0

    def increment(self):
        if self.value < self.max_value:
            self.value += 1
        else:
            print("Error: Counter has reached its maximum value!")

    def reset(self):
        self.value = 0

    def get_value(self):
        return self.value
```

8. Add a function to the object classes, including the base, that tests if a given set of $x$ and $y$ coordinates are inside of the object. You'll have to think through how to determine if a set of coordinates are inside an object for each object type.

```python
# Solution
class Shape:
    def area(self):
        pass

    def perimeter(self):
        pass

    def get_points(self):
        pass

    def is_inside(self, x, y):
        pass


# Rectangle class
class Rectangle(Shape):
    def __init__(self, x, y, width, length):
        self.__x = x
        self.__y = y
        self.__width = width
        self.__length = length

    def is_inside(self, x, y):
        if self.__x <= x <= self.__x + self.__width and self.__y <= y <= self.__y + self.__length:
            return True
        return False


# Circle class
class Circle(Shape):
    def __init__(self, x, y, radius):
        self.__x = x
        self.__y = y
        self.__radius = radius

    def is_inside(self, x, y):
        dx = x - self.__x
        dy = y - self.__y
        if dx * dx + dy * dy <= self.__radius * self.__radius:
            return True
        return False


# Triangle class (I'm doing a right triangle)
class Triangle(Shape):
    def __init__(self, x, y, side1, side2, side3):
        self.__x = x
        self.__y = y
        self.__side1 = side1
        self.__side2 = side2
        self.__side3 = side3

    def is_inside(self, x, y):
        vertex1 = (self.__x, self.__y)
        vertex2 = (self.__x + self.__side1, self.__y)
        vertex3 = (self.__x, self.__y + self.__side2)

        def triangle_area(a, b, c):
            return 0.5 * abs(
                a[0]*(b[1]-c[1]) +
                b[0]*(c[1]-a[1]) +
                c[0]*(a[1]-b[1])
            )

        full_area = triangle_area(vertex1, vertex2, vertex3)
        area1 = triangle_area((x, y), vertex2, vertex3)
        area2 = triangle_area(vertex1, (x, y), vertex3)
        area3 = triangle_area(vertex1, vertex2, (x, y))

        if full_area == area1 + area2 + area3:
            return True
        return False
```

```python
# Test Function
rect = Rectangle(1, 2, 5, 10)
circle = Circle(3, 4, 5)
triangle = Triangle(0, 0, 3, 4, 5)

print("Rectangle (3,5)?", rect.is_inside(3, 5))
print("Circle (7,4)?", circle.is_inside(7, 4))
print("Triangle (1,1)?", triangle.is_inside(1, 1))
```

9. Add a function in the base class of the object classes that returns true/false testing that the object overlaps with another object.

```python
# Solution
class Shape:
    def area(self):
        pass

    def perimeter(self):
        pass

    def get_points(self):
        pass

    def is_inside(self, x, y):
        pass

    def get_bounds(self):
        pass

    def overlaps(self, other):
        # Check if any point of this object is inside the other
        points = self.get_points()
        for px, py in points:
            if other.is_inside(px, py):
                return True
        # Also check if any point of the other object is inside this object
        other_points = other.get_points()
        for px, py in other_points:
            if self.is_inside(px, py):
                return True
        return False


# Rectangle class
class Rectangle(Shape):
    def __init__(self, x, y, width, length):
        self.__x = x
        self.__y = y
        self.__width = width
        self.__length = length

    def is_inside(self, x, y):
        return self.__x <= x <= self.__x + self.__width and self.__y <= y <= self.__y + self.__length

    def get_bounds(self):
        return (self.__x, self.__y, self.__x + self.__width, self.__y + self.__length)

    def get_points(self):
        corners = [
            (self.__x, self.__y),
            (self.__x + self.__width, self.__y),
            (self.__x + self.__width, self.__y + self.__length),
            (self.__x, self.__y + self.__length)
        ]
        points = []
        for i in range(16):
            points.append(corners[i % 4])
        return points


# Circle class
class Circle(Shape):
    def __init__(self, x, y, radius):
        self.__x = x
        self.__y = y
        self.__radius = radius

    def is_inside(self, x, y):
        dx = x - self.__x
        dy = y - self.__y
        return dx*dx + dy*dy <= self.__radius*self.__radius

    def get_bounds(self):
        return (self.__x - self.__radius, self.__y - self.__radius,
                self.__x + self.__radius, self.__y + self.__radius)

    def get_points(self):
        points = []
        for i in range(16):
            if i % 4 == 0:
                points.append((self.__x + self.__radius, self.__y))
            elif i % 4 == 1:
                points.append((self.__x, self.__y + self.__radius))
            elif i % 4 == 2:
                points.append((self.__x - self.__radius, self.__y))
            else:
                points.append((self.__x, self.__y - self.__radius))
        return points


# Triangle class (right triangle simplification)
class Triangle(Shape):
    def __init__(self, x, y, side1, side2, side3):
        self.__x = x
        self.__y = y
        self.__side1 = side1
        self.__side2 = side2
        self.__side3 = side3

    def is_inside(self, x, y):
        vertex1 = (self.__x, self.__y)
        vertex2 = (self.__x + self.__side1, self.__y)
        vertex3 = (self.__x, self.__y + self.__side2)

        def triangle_area(a, b, c):
            return 0.5 * abs(
                a[0]*(b[1]-c[1]) +
                b[0]*(c[1]-a[1]) +
                c[0]*(a[1]-b[1])
            )

        full_area = triangle_area(vertex1, vertex2, vertex3)
        area1 = triangle_area((x, y), vertex2, vertex3)
        area2 = triangle_area(vertex1, (x, y), vertex3)
        area3 = triangle_area(vertex1, vertex2, (x, y))

        return full_area == area1 + area2 + area3

    def get_bounds(self):
        return (self.__x, self.__y, self.__x + self.__side1, self.__y + self.__side2)

    def get_points(self):
        vertices = [
            (self.__x, self.__y),
            (self.__x + self.__side1, self.__y),
            (self.__x, self.__y + self.__side2)
        ]
        points = []
        for i in range(16):
            points.append(vertices[i % 3])
        return points
```

```python
rect = Rectangle(1, 1, 4, 4)
circle = Circle(2, 2, 1)
triangle = Triangle(0, 0, 3, 4, 5)

print("Rectangle overlaps Circle?", rect.overlaps(circle))
print("Rectangle overlaps Triangle?", rect.overlaps(triangle))
print("Circle overlaps Triangle?", circle.overlaps(triangle))
print("Rectangle is_inside (3,3)?", rect.is_inside(3,3))
print("Circle is_inside (0,0)?", circle.is_inside(0,0))
print("Triangle is_inside (1,1)?", triangle.is_inside(1,1))
```

10. Copy the `Canvas` class from lecture to in a python file creating a `paint` module. Copy your classes from above into the module and implement paint functions. Implement a `CompoundShape` class. Create a simple drawing demonstrating that all of your classes are working.

```python
# Solution
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        if 0 <= row < self.height and 0 <= col < self.width:
            self.data[row][col] = char

    def line(self, x1, y1, x2, y2, char='*'):
        dx = x2 - x1
        dy = y2 - y1
        steps = max(abs(dx), abs(dy))
        if steps == 0:
            self.set_pixel(y1, x1, char)
            return
        x_inc = dx / steps
        y_inc = dy / steps
        x = x1
        y = y1
        for i in range(steps + 1):
            self.set_pixel(int(round(y)), int(round(x)), char)
            x += x_inc
            y += y_inc

    def display(self):
        for row in self.data:
            print("".join(row))

    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]

# Edited the Shape class
class Shape:
    def paint(self, canvas):
        pass


class Rectangle(Shape):
    def __init__(self, x, y, width, length):
        self.__x = x
        self.__y = y
        self.__width = width
        self.__length = length

    def paint(self, canvas, char='*'):
        # draw edges
        for i in range(self.__width):
            canvas.set_pixel(self.__y, self.__x + i, char)
            canvas.set_pixel(self.__y + self.__length - 1, self.__x + i, char)
        for i in range(self.__length):
            canvas.set_pixel(self.__y + i, self.__x, char)
            canvas.set_pixel(self.__y + i, self.__x + self.__width - 1, char)

class Circle(Shape):
    def __init__(self, x, y, radius):
        self.__x = x
        self.__y = y
        self.__radius = radius

    def paint(self, canvas, char='*'):
        for dy in range(-self.__radius, self.__radius + 1):
            for dx in range(-self.__radius, self.__radius + 1):
                if dx*dx + dy*dy <= self.__radius*self.__radius:
                    canvas.set_pixel(self.__y + dy, self.__x + dx, char)

class Triangle(Shape):
    def __init__(self, x, y, side1, side2, side3):
        self.__x = x
        self.__y = y
        self.__side1 = side1
        self.__side2 = side2
        self.__side3 = side3

    def paint(self, canvas, char='*'):
        # we're doing a right triangle
        canvas.line(self.__x, self.__y, self.__x + self.__side1, self.__y, char)
        canvas.line(self.__x, self.__y, self.__x, self.__y + self.__side2, char)
        canvas.line(self.__x + self.__side1, self.__y, self.__x, self.__y + self.__side2, char)


class CompoundShape(Shape):
    def __init__(self, shapes):
        self.shapes = shapes

    def paint(self, canvas, char='*'):
        for shape in self.shapes:
            shape.paint(canvas, char)
```

```python
# Test Function
canvas = Canvas(40, 20)

rect = Rectangle(1, 1, 10, 5)
circle = Circle(25, 5, 4)
triangle = Triangle(5, 12, 8, 4, 5)

compound = CompoundShape([rect, circle, triangle])
compound.paint(canvas)
canvas.display()
```

11. Create a `RasterDrawing` class. Demonstrate that you can create a drawing made of several shapes, paint the drawing, modify the drawing, and paint it again. 

```python
class Shape:
    def __init__(self, name=""):
        self.name = name

    def paint(self, canvas):
        pass

class RasterDrawing:
    def __init__(self):
        self.shapes = dict()
        self.shape_names = list()
        
    def add_shape(self, shape):
        if not hasattr(shape, "name") or shape.name == "":
            shape.name = self.assign_name()
        self.shapes[shape.name] = shape
        self.shape_names.append(shape.name)

    def update(self, canvas):
        canvas.clear_canvas()
        self.paint(canvas)
        
    def paint(self, canvas):
        for shape_name in self.shape_names:
            self.shapes[shape_name].paint(canvas)
            
    def assign_name(self):
        name_base = "shape"
        name = name_base + "_0"
        i = 1
        while name in self.shapes:
            name = name_base + "_" + str(i)
            i += 1
        return name
```

```python
# Solution
class Shape:
    def __init__(self, name=""):
        self.name = name
    def paint(self, canvas):
        pass

class Rectangle(Shape):
    def __init__(self, x, y, width, length, name=""):
        super().__init__(name)
        self.x = x
        self.y = y
        self.w = width
        self.h = length
    def paint(self, canvas, char='*'):
        for i in range(self.w):
            canvas.set_pixel(self.y, self.x + i, char)
            canvas.set_pixel(self.y + self.h - 1, self.x + i, char)
        for i in range(self.h):
            canvas.set_pixel(self.y + i, self.x, char)
            canvas.set_pixel(self.y + i, self.x + self.w - 1, char)
    def __repr__(self):
        return f"Rectangle({self.x},{self.y},{self.w},{self.h},name={repr(self.name)})"

class Circle(Shape):
    def __init__(self, x, y, radius, name=""):
        super().__init__(name)
        self.x = x
        self.y = y
        self.r = radius
    def paint(self, canvas, char='*'):
        for dy in range(-self.r, self.r + 1):
            for dx in range(-self.r, self.r + 1):
                if dx*dx + (2*dy)*(2*dy) <= self.r*self.r:
                    canvas.set_pixel(self.y + dy, self.x + dx, char)
    def __repr__(self):
        return f"Circle({self.x},{self.y},{self.r},name={repr(self.name)})"

class Triangle(Shape):
    def __init__(self, x, y, side1, side2, side3, name=""):
        super().__init__(name)
        self.x = x
        self.y = y
        self.w = side1
        self.h = side2
        self.s3 = side3
    def paint(self, canvas, char='*'):
        canvas.line(self.x, self.y, self.x + self.w, self.y, char)
        canvas.line(self.x, self.y, self.x, self.y + self.h, char)
        canvas.line(self.x + self.w, self.y, self.x, self.y + self.h, char)
    def __repr__(self):
        return f"Triangle({self.x},{self.y},{self.w},{self.h},{self.s3},name={repr(self.name)})"

class RasterDrawing:
    def __init__(self):
        self.shapes = dict()
        self.shape_names = list()
    def add_shape(self, shape):
        if shape.name == "":
            shape.name = self.assign_name()
        self.shapes[shape.name] = shape
        self.shape_names.append(shape.name)
    def assign_name(self):
        i = 0
        name = f"shape_{i}"
        while name in self.shapes:
            i += 1
            name = f"shape_{i}"
        return name
    def paint(self, canvas):
        for name in self.shape_names:
            self.shapes[name].paint(canvas)
    def __repr__(self):
        s = ""
        for name in self.shape_names:
            s += f"rd.add_shape({repr(self.shapes[name])})\n"
        return s
    def save(self, filename):
        with open(filename, "w") as f:
            f.write(self.__repr__())

def raster_loader(filename):
    rd = RasterDrawing()
    with open(filename, "r") as f:
        content = f.read()
    exec(content)
    return rd
```
